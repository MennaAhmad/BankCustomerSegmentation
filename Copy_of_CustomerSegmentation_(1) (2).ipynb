```python
def categorize_time(hour, minute):
    total_minutes = hour * 60 + minute

    if 5 * 60 <= total_minutes < 12 * 60:
        return 'Morning'
    elif 12 * 60 <= total_minutes < 17 * 60:
        return 'Noon'
    elif 17 * 60 <= total_minutes < 20 * 60:
        return 'Evening'
    else:
        return 'Night'
```

#K-means

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import pandas as pd

# Select relevant features for clustering
features = ['CustAccountBalance', 'TransactionAmount (INR)', 'Age']

# Preprocess the data (standard scaling)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])


```

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['BudgetCategory'] = kmeans.fit_predict(X_scaled)

```

```python
# Display the number of customers in each class (BudgetCategory)
cluster_counts = df['BudgetCategory'].value_counts()

print("Number of customers in each BudgetCategory (Cluster):")
print(cluster_counts)

# For clarity, you can print the clusters as Low, Medium, and High Budget based on the cluster index
df['BudgetCategoryLabel'] = df['BudgetCategory'].map({
    0: 'Low Budget',
    1: 'Medium Budget',
    2: 'High Budget'
})

print("Number of customers in each BudgetCategory (Low, Medium, High):")
print(df['BudgetCategoryLabel'].value_counts())
```

#Classification

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, mean_squared_error, r2_score

# Features and target for classification and regression
X = df[['CustAccountBalance', 'TransactionAmount (INR)', 'Age']]
y_class = df['BudgetCategory']  # Target for classification
y_reg = df['TransactionAmount (INR)']  # Target for regression

# Train-test split for classification
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X, y_class, test_size=0.2, random_state=42)

# Train-test split for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.2, random_state=42)

# RandomForest Classifier
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train_class, y_train_class)
class_predictions = rf_classifier.predict(X_test_class)


```

```python
# RandomForest Regressor
rf_regressor = RandomForestRegressor(random_state=42)
rf_regressor.fit(X_train_reg, y_train_reg)
reg_predictions = rf_regressor.predict(X_test_reg)

# Evaluate the RandomForest Classifier
print("RandomForest Classifier Report:")
print(classification_report(y_test_class, class_predictions))

# Evaluate the RandomForest Regressor
print("RandomForest Regressor Report:")
print(f'RÂ² Score: {r2_score(y_test_reg, reg_predictions)}')
print(f'Mean Squared Error: {mean_squared_error(y_test_reg, reg_predictions)}')
```
